```python
import pandas as pd
import numpy as np

# Create a data frame with 25 rows
rows = 25
df = pd.DataFrame({'A': np.random.randn(rows),
                   'B': np.random.randn(rows),
                   'C': np.random.randn(rows),
                   'D': np.random.randn(rows),
                   'E': np.random.randn(rows)})

# Add missing values
df.iloc[10:15, 0:3] = np.nan
df.iloc[20:25, 3:5] = np.nan

# Add non-finite values
df.iloc[0:5, 0:2] = np.inf
df.iloc[5:10, 2:4] = -np.inf

print(df.head(25))
```
